```python
##importing the dataset
import pandas as pd
df=pd.read_csv(r"C:\Users\USER\Downloads\archive (6)\gdp_per_capita_growth.csv")
df
```

```python
##removing 1960 since all values are null
df.drop('1960',axis=1)
```

```python
##checking for missing values
df.isna().sum()
```

```python
##filling the null values uding backfill methos
df=df.fillna(method='bfill')
df
```

```python
df.isna().sum()
```

```python
##creating a sub-dataset from 2015 to 2020
data=df[['2015','2016','2017','2018','2019','2020']]
data
```

```python
##independent variables
x=data.drop('2020',axis=1)
x
```

```python
##dependent variable
y=df['2020']
y
```

```python
##scatter plot of the independent variables against the dependent variable
import matplotlib.pyplot as plt
for i in x.columns:
    plt.scatter(x[i],y,s=2)
    plt.ylabel('2020')
    plt.xlabel('2015-2019')
    
##legend and title
plt.legend(x.columns)
plt.title('Scatter plot of 2015-2019 against 2020',color='magenta')
```

```python
##splitting the data into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=24)
```

```python
##calling linear regression
from sklearn.linear_model import LinearRegression
linear_regression=LinearRegression()
```

```python
##fitting the train data
linear_regression.fit(xtrain,ytrain)
```

```python
##intercept
linear_regression.intercept_
```

```python
##coefficients
linear_regression.coef_
```

```python
##testing the model
ypred_train=linear_regression.predict(xtrain)
```

```python
##comparing the predicted values and the actual values in training data
for p, e in zip(ypred_train[::], ytrain[::]):
    print(f'predicted:{p:.2f}, actual: {e:.2f}')
```

```python
##scatter plot for comparison of actual and predicted values
plt.scatter(ytrain,ypred_train)
```

```python
##calculating r-squared
from sklearn.metrics import r2_score
r2_score(ytrain,ypred_train)
```

```python
###model has low r-score value of 3.96%
```

```python
##predicting y-value(2020) using the test data
ypred_test=linear_regression.predict(xtest)
```

```python
##comparing the predicted y values to actual values
for p, e in zip(ypred_test[::], ytest[::]):
    print(f'predicted:{p:.2f}, actual: {e:.2f}')
```

```python
##scatter plot for comparison
plt.scatter(ytest,ypred_test)
```

```python
##calculating r-squared values
r2_score(ytest,ypred_test)
```

```python
##model has low r-squared value of 24%
```
