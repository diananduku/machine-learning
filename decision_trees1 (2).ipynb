```python
##importing the dataset
import pandas as pd
data=pd.read_csv(r"C:\Users\USER\Downloads\archive (11)\diabetes_prediction_dataset.csv")
data
```

```python
##checking for missing values
data.isna().sum()
```

```python
##converting categorical columns to numeric
data['gender']=data['gender'].astype('category')
data['gender']=data['gender'].cat.codes
data['smoking_history']=data['smoking_history'].astype('category')
data['smoking_history']=data['smoking_history'].cat.codes
data
```

```python
##selecting the input
input=data.drop('diabetes',axis=1)
input
```

```python
target=data['diabetes']
target
```

```python
##splitting data into 70% training and 30% testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.3,random_state=1)
```

```python
##decision tree 
from sklearn import tree
model=tree.DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
model=model.fit(x_train,y_train)
model
```

```python
##testing accuracy of the model using training data
model.score(x_train,y_train)
```

```python
##testing the model
test_prediction=model.predict(x_test)
test_prediction
```

```python
## testing the accuracy of the model using test data
model.score(x_test,y_test)
```
